```python
# імпорт бібліотек

import matplotlib.pyplot as plt
import numpy as np
import sympy as sp
```

```python
def check_len(obj):
    try:
        return len(obj)
    except TypeError:
        return 0
```

```python
# функція для побудови графіків функції та її похідної

def derivative_viz(x,func,x_val,y_val):

    # визначення даних
    derivative_func = sp.diff(func,x)
    df_lambda = sp.lambdify(x, derivative_func, 'numpy')
    df_val = df_lambda(x_val)
    
    # перевірка, чи є похідна константою
    if check_len(df_val) == 0:
        df_val = [df_val]*len(x_val)

    print("Похідна функції ",func,":", derivative_func)
    
    # візуалізація
    
    plt.figure(figsize = (6,3))
    plt.plot(x_val, y_val, label = 'функція f(x)=' + str(func), color = "firebrick")
    plt.plot(x_val, df_val, label = 'похідна f`(x)=' + str(derivative_func), color = "olivedrab",ls = '--')
    
    plt.grid()
    plt.xlabel('x')
    plt.ylabel('y')
    plt.legend()
    plt.title('Функція f(x)=' + str(func) + "та її похідна")
    plt.show()
```

Існує пряма лінійна залежність між кількістю годин навчання та оцінками, тобто оцінки зростають з ростом часу занять.

2. Використовуючи Python і `SymPy`, обчисліть похідні для функцій $ f(x) = \sin(x) $ та $ f(x) = \ln(x) $. Зобразіть ці функції та їхні похідні за допомогою `matplotlib`.



```python
# визначення даних

x = sp.symbols('x')
func = sp.sin(x)
x2_values = np.linspace(0,10,100)
y2_values = np.sin(x2_values)

# візуалізація
derivative_viz(x, func, x2_values, y2_values)

```

```python
# визначення даних

x = sp.symbols('x')
func = sp.log(x)
x3_values = np.linspace(1,10,100)
y3_values = np.log(x3_values)

# візуалізація

derivative_viz(x, func, x3_values, y3_values)
```
