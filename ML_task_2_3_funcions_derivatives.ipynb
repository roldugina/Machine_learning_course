```python
# імпорт бібліотек

import matplotlib.pyplot as plt
import numpy as np
import sympy as sp
```

```python
# функція для побудови графіків функції та її похідної

def derivative_viz(x,func,x_val,y_val,title):

    derivative_func = sp.diff(func,x)
    df_lambda = sp.lambdify(x, derivative_func, 'numpy')
    df_val = df_lambda(x_val)
    
    print("Похідна функції ",func,":", derivative_func)
        
    plt.figure(figsize = (6,3))
    plt.plot(y_val, label = 'функція f(x)=' + str(func), color = "firebrick")
    plt.plot(df_val, label = 'похідна f`(x)=' + str(derivative_func), color = "olivedrab",ls = '--')

    plt.grid()
    plt.xlabel('x')
    plt.ylabel('y')
    plt.legend()
    plt.title(title)
    plt.show()
```

```python
# обчислення похідної

x = sp.symbols('x')
func = 10*x-5
```

```python
# визначення вихідних даних

x_values = np.linspace(0,10,11).astype(int)
y_values = 10*x_values - 5

# побудова графіків
derivative_viz(x, func, x_values, y_values, 'Залежність оцінки від кількості годин, витрачених на навчання')
```

2. Використовуючи Python і `SymPy`, обчисліть похідні для функцій $ f(x) = \sin(x) $ та $ f(x) = \ln(x) $. Зобразіть ці функції та їхні похідні за допомогою `matplotlib`.



```python
x = sp.symbols('x')
func = sp.sin(x)
x_values = np.linspace(0,10,100)
y_values = np.sin(x_values)

derivative_viz(x, func, x_values, y_values, 'Функція sin(x) та її похідна')

```

```python
x = sp.symbols('x')
func = sp.ln(x)
x_values = np.linspace(1,10,100)
y_values = np.sin(x_values)

derivative_viz(x, func, x_values, y_values, 'Функція sin(x) та її похідна')
```
